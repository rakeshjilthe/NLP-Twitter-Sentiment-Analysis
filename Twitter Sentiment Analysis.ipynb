### Data Processing

```python
def data_processing(text):
    text = text.lower()                                                                 #Lowercase conversion
    text = re.sub(r"https\S+|www\S+https\S+", '', text, flags = re.MULTILINE)           #remove urls in the tweets
    text = re.sub(r'\@w+|\#','',text)                                                   #removing # and punctuations
    text = re.sub(r'[^\w\s]','',text)
    text_token = word_tokenize(text)                                                    #removing the stopwords
    filtered_text = [w for w in text_token if not w in stop_words]
    return " ".join(filtered_text)
```

```python
text_df.text = text_df['text'].apply(data_processing)
```

```python
text_df = text_df.drop_duplicates('text')                                               # removing dublicates data and text
```

### Performing Porter Stemming

```python
stemmer = PorterStemmer()

def stemming(data):
    text = [stemmer.stem(word) for word in data]
    return data
```

```python
text_df['text'] = text_df['text'].apply(lambda x: stemming(x))
```

#### Let's use TextBlob to calculate the polarity of the sentences

```python
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
text_df['polarity'] = text_df['text'].apply(polarity)
```

```python
text_df.head(10)
```

Now, we can polarity column has been added to the dataframe.

#### We can also add sentiment column to the dataframe.

```python
def sentiment(label):
    if label < 0:
        return "Negative"
    elif label == 0:
        return "Neutral"
    elif label > 0:
        return "Positive"
```

```python
text_df['sentiment'] = text_df['polarity'].apply(sentiment)
```

### Now let's create the Bigram model and perform Count Vectorization

```python
vect = CountVectorizer(ngram_range=(1,2)).fit(text_df['text'])
```

```python
feature_names = vect.get_feature_names()
print("Number of features: {}\n".format(len(feature_names)))
print("First 20 features:\n {}".format(feature_names[:20]))
```

Now, we have vectorized the model we can proceed further to build the model.

```python
X = text_df['text']
Y = text_df['sentiment']
X = vect.transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
print('Size of x_train:', (x_train.shape))
print('Size of y_train:', (y_train.shape))
print('Size of x_test:', (x_test.shape))
print('Size of y_test:', (y_test.shape))
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
logreg_pred = logreg.predict(x_test)
logreg_acc = accuracy_score(logreg_pred, y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc * 100))
```

```python
print(confusion_matrix(y_test, logreg_pred))
print("\n")
print(classification_report(y_test, logreg_pred))
```

### We can display confusion matrix in different way by using Confusion matrix display, let us see.

```python
style.use('classic')
cm = confusion_matrix(y_test, logreg_pred, labels = logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = logreg.classes_)
disp.plot()
```

```python
from sklearn.model_selection import GridSearchCV
```
